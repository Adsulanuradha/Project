**Chi-Square Test (for checking association between independent varible )**

```python
from sklearn.preprocessing import LabelEncoder
for col in data.columns:
  le=LabelEncoder()
  data[col]=le.fit_transform(data[col])
data.head()
```

```python
from sklearn.feature_selection import chi2
x=data.drop(columns=['category_Accessories'],axis=1)
y=data['category_Accessories']
chi_scores=chi2(x,y)
```

```python
chi_scores
```

```python
chi_values=pd.Series(chi_scores[0],index=x.columns)
chi_values.sort_values(ascending=False,inplace=True)
chi_values.plot.bar()
#high chi values means high importance
```
